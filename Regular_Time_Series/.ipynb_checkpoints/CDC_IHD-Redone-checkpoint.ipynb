```python
X_sm.corr()
```

```python
y_pred = sm_pois.predict(X_test)

def r2(a,b):
    SSE = 0
    SST = 0
    for i, j in zip(a,b):
        SSE += (j - i)**2
        SST += (i - a.mean())**2 
    return 1 - SSE/SST

r2(y_test,y_pred)
```
